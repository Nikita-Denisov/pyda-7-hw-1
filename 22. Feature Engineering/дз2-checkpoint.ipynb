```python
names = [x for x in data]
df_norm = pd.DataFrame(sklearn.preprocessing.MinMaxScaler(feature_range=(-1,1)).fit_transform(data), columns=data.columns)
```

Как мы видим, набольшая корреляция с MEDV у признаков LSTAT(-0,85, обратная), RM(0,63) и INDUS (0,58)

```python
X = df_norm.drop(['MEDV'], axis=1)
y = df_norm['MEDV']
```

```python
stand_scaler = StandardScaler()
X_stand = stand_scaler.fit_transform(X)
X_stand = pd.DataFrame(X_stand, columns=X.columns)
X_stand.head(5)
```

```python
minmax_scaler = MinMaxScaler()
X_mm = minmax_scaler.fit_transform(X)
X_mm = pd.DataFrame(X_mm, columns=X.columns)
X_mm.head(5)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
X_stand_train, X_stand_test, y_stand_train, y_stand_test = train_test_split(X_stand, y, test_size=0.2, random_state=123)
X_mm_train, X_mm_test, y_mm_train, y_mm_test = train_test_split(X_mm, y, test_size=0.2, random_state=123)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

Функция для обучения подели:

```python
def train_model(columns):
    estimator = LinearRegression()

    estimator.fit(X_train[columns], y_train)
    prediction = estimator.predict(X_test[columns])
    print('As is MAE: {}'.format(mean_absolute_error(y_test, prediction)))

    estimator.fit(X_stand_train[columns], y_stand_train)
    prediction = estimator.predict(X_stand_test[columns])
    print('Standart scaller MAE: {}'.format(mean_absolute_error(y_stand_test, prediction)))

    estimator.fit(X_mm_train[columns], y_mm_train)
    prediction = estimator.predict(X_mm_test[columns])
    print('MinMax scaller MAE: {}'.format(mean_absolute_error(y_mm_test, prediction)))

train_model(X.columns)
```

Мы получили результат при обучении модели со всем признаками.

Теперь поробуем пообучать модель по каждому признаку

```python
for i in X.columns:
    print('Ошибка при обучении модели по признаку',i,':')
    print(train_model([i]))
    print()
```

Ошибка меньше всего при обучении с параметрами RM и LSTAT (по 0.22)

Попробуем улучшить модель, используя среднюю, медиану, минимальное значение 

```python
MEDV_pivot = df_norm.pivot_table(values=['RM', 'LSTAT'], index='MEDV', aggfunc=['min', 'count', 'mean', 'median'])
```

```python
MEDV_data = df_norm['MEDV'].apply(lambda x: MEDV_pivot.loc[x].reset_index(drop=True))
MEDV_data.columns = ['min_RM', 'min_LSTAT',
                    'count_RM', 'count_LSTAT',
                    'mean_RM', 'mean_LSTAT',
                    'median_RM', 'median_LSTAT']
MEDV_data.head(5)
```

```python
df_norm = df_norm.join(MEDV_data)
df_norm.head(5)
```

```python
X = df_norm.drop(['MEDV'], axis=1)
y = df_norm['MEDV']
```

```python
stand_scaler = StandardScaler()
X_stand = stand_scaler.fit_transform(X)
X_stand = pd.DataFrame(X_stand, columns=X.columns)
X_stand.head(5)
```

```python
minmax_scaler = MinMaxScaler()
X_mm = minmax_scaler.fit_transform(X)
X_mm = pd.DataFrame(X_mm, columns=X.columns)
X_mm.head(5)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
X_stand_train, X_stand_test, y_stand_train, y_stand_test = train_test_split(X_stand, y, test_size=0.2, random_state=123)
X_mm_train, X_mm_test, y_mm_train, y_mm_test = train_test_split(X_mm, y, test_size=0.2, random_state=123)
```

Обучим модель только только на наших новых показателях и посмотрим результат.

```python
train_model(['min_RM','min_LSTAT','count_RM','count_LSTAT','mean_RM','mean_LSTAT','median_RM','median_LSTAT'])
```
